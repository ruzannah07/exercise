#### total keuntunganSemata tiap productline yg sold (status = on hold & shipped)
##### product line akan kita anggap sebagai brand.

```python
engine.execute("""
CREATE VIEW Realita AS
SELECT productLine, SUM(priceEach - buyPrice) as keuntunganSemata
FROM products JOIN orderdetails using (productCode)
JOIN orders using (orderNumber)
WHERE status IN ('on hold','shipped') AND priceEach > buyPrice
GROUP BY productLine
""")
```

```python
#target keuntngan MSRP*quantity
```

```python
engine.execute("""
CREATE VIEW Ekspektasi AS
SELECT productLine, SUM(MSRP-buyPrice) as keuntunganTarget
FROM products JOIN orderdetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE status IN ('on hold','shipped')
GROUP BY productLine
""")
```

#### berapa persen kita mendekati target keuntungan????

```python
dataKeuntungan = run("""SELECT productLine,keuntunganSemata, 
keuntunganTarget,(keuntunganSemata/keuntunganTarget)*100 as persentase
FROM realita JOIN ekspektasi USING (productLine)
ORDER BY persentase DESC
""")
```

### SOAL - Many to many relationship
1. List products sold by order date.
2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
3. List the names of customers and their corresponding order number 
   where a particular order from that customer has a value greater than $25,000?
4. Are there any products that appear on all orders?
5. List the names of products sold at less than 80% of the MSRP.
6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)
7. List the products ordered on a Monday.
8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
#2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT productName, orderDate
FROM products JOIN orderDetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE productName = '1940 Ford Pickup Truck'
ORDER BY orderDate DESC
""")
```

```python
#3. List the names of customers and their corresponding order number 
#   where a particular order from that customer has a value greater than $25,000?
run("""
SELECT customerName, orderNumber , SUM(priceEach*quantityOrdered)as totbelanjaan
FROM customers JOIN orders USING (customerNumber)
JOIN orderDetails USING (orderNumber)
GROUP BY orderNumber
HAVING totbelanjaan > 25000
ORDER BY customerName
""")    
    
```

```python
#4. Are there any products that appear on all orders?
run("""
SELECT productName, COUNT(*)as total
FROM products JOIN orderDetails using (productCode)
GROUP BY productName
HAVING total = (SELECT COUNT(*) FROM orders)
""")
```

```python
#5. List the names of products sold at less than 80% of the MSRP.
run("""
SELECT orderNumber, productCode, priceEach, MSRP
FROM products JOIN orderDetails USING (productCode)
WHERE priceEach < 0.8*MSRP
""")
```

```python
#6. Reports those products that have been sold with a markup (kenaikan harga) of 100% or more 
#(i.e.,  the priceEach is at least twice the buyPrice)

run("""
SELECT orderNumber, productName, priceEach, buyPrice
FROM products JOIN orderDetails USING (productCode)
WHERE priceEach >= 2*buyPrice
""")
```

```python
#PENTING
#7. List the products ordered on a Monday.
run("""
SELECT productName, DAYNAME(orderDate)
FROM products JOIN orderDetails USING (productCode)
JOIN orders USING (orderNumber)
WHERE DAYOFWEEK(orderDate) = 2
""")

#dayname() hari 'sunday', 'monday'
#dayofweek() hari dlm bentuk angka 1=sunday
#year() tahun
#month() bulan ke
#date() dari format thn-bln-tgl jam:menit:detik, yg diambil date without time
#DAY() 1-30 tanggalan
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT productName, status
FROM products JOIN orderDetails using (productCode)
JOIN orders using (ordernumber)
WHERE status = 'On hold'
""")
```

```python
year = run("select distinct year(shippedDate) as tahun from orders where shippedDate is not null")
year = list(year['tahun'])

year = run("select distinct year(shippedDate) as tahun from orders where shippedDate is not null")
year = list(year['tahun'])
years = []
for i in year:
    years.append(str(i))
years
```

```python
for i in years:
    sql = "SELECT MONTH(shippedDate)as bulan , SUM(priceEach) as totalPenjualan FROM orders JOIN orderDetails USING (orderNumber) WHERE YEAR(shippedDate)='"+ i +"' GROUP BY bulan" 
    datanya = run(sql)
    print("===========================")
    print("TOTAL PENJUALAN TAHUN "+ i) 
    print(datanya)
    print("\n")
    
#brarti shippedDate is not null, alias, sudah dikirim
```
